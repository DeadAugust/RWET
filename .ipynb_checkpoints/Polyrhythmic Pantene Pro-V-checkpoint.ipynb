# Polyrhythmic Pantene Pro-V

## New Poetic Form

### Polyrhythmic Hybrid between Pantoum and a Sestina

### pantoum

poets.org definition: "The pantoum originated in Malaysia in the fifteenth-century as a short folk poem, typically made up of two rhyming couplets that were recited or sung. However, as the pantoum spread, and Western writers altered and adapted the form, the importance of rhyming and brevity diminished. The modern pantoum is a poem of any length, composed of four-line stanzas in which the second and fourth lines of each stanza serve as the first and third lines of the next stanza. The last line of a pantoum is often the same as the first."

*Incanations, Rituals, Prayers*

interwoven lines with structure similar to
- ABCD
- BEDF
- EGFH
- GCHA

### sestina

poets.org definition: "The sestina follows a strict pattern of the repetition of the initial six end-words of the first stanza through the remaining five six-line stanzas, culminating in a three-line envoi. The lines may be of any length, though in its initial incarnation, the sestina followed a syllabic restriction. The form is as follows, where each numeral indicates the stanza position and the letters represent end-words:

1. ABCDEF
2. FAEBDC
3. CFDABE
4. ECBFAD
5. DEACFB
6. BDFECA
7. (envoi) ECA or ACE

The envoi, sometimes known as the tornada, must also include the remaining three end-words, BDF, in the course of the three lines so that all six recurring words appear in the final three lines. In place of a rhyme scheme, the sestina relies on end-word repetition to effect a sort of rhyme."

```python
# now combine and remove periods

gemini = []
for phrase in proverbs:
    phraseClean = phrase.replace('.', '')
    gemini.append(phraseClean)
for phrase in harvard:
    phraseClean = phrase.replace('.', '')
    gemini.append(phraseClean)
print(gemini)
```

### then we'll randomly select six ending words for the sestina

```python
#have to remove punct. for rhymes to work
sestinaSeeds = []
for i in range(6):
    seed = random.choice(gemini).split()
    seedClean = seed[-1]
    sestinaSeeds.append(seedClean)
print(sestinaSeeds)
```

### then generate a list of words that rhyme with each of those

```python
# have to add the original word back!!!

sestinaRhymes = []
for seeds in sestinaSeeds:
    allSeeds = []
    allSeeds = pronouncing.rhymes(seeds)
    allSeeds.append(seeds)
    sestinaRhymes.append(allSeeds)
#     sestinaRhymes.append(pronouncing.rhymes(seeds))
print(sestinaRhymes)
```

```python
# sestinaAs = [for rhymes in sestinaAwords if sestinaAwords == ]

#why aren't all the phrases showing up?? 
#the original phrase from which the seed was extracted isn't appearing
sestinaPhrases = []

#    sestinaPhrases.append([phrase for word in rhymes if phrase[-1] == word])

for rhymes in sestinaRhymes:
    phraseList = []
    for word in rhymes:
        for phrase in gemini:
            if phrase[-1] == word:
                phraseList.append(phrase)
    if len(phraseList) >= 1:
        sestinaPhrases.append(phraseList)

print(sestinaPhrases)

```
