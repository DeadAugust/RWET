```python
#from Allison's notebook

def add_to_model(model, n, seq):
    # make a copy of seq and append None to the end
    seq = list(seq[:]) + [None]
    for i in range(len(seq)-n):
        # tuple because we're using it as a dict key!
        gram = tuple(seq[i:i+n])
        next_item = seq[i+n]            
        if gram not in model:
            model[gram] = []
        model[gram].append(next_item)

def markov_model(n, seq):
    model = {}
    add_to_model(model, n, seq)
    return model
```

```python
corpus_markov_model_2 = markov_model(2, open('txt/JacksonCorpus_Apr19.txt').read().split())
corpus_markov_model_3 = markov_model(3, open('txt/JacksonCorpus_Apr19.txt').read().split())
```
